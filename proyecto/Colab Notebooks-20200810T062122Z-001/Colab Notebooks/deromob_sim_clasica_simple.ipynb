```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 0.01
t_dur = 20

n_steps = int(t_dur/delta)
print(n_steps)

t=np.linspace(0,t_dur,n_steps+1)
a = np.ones_like(t)*0.1
#a=-np.sin(t)


#a[a<0]=0
plt.plot(t,a)
```

Usando ecuaciones de diferencia:
[eq1] v(t+dt) = v(t)+dt*a(t)+ otros
[eq2] s(t+dt)= s(t)+dt*v(t)+otros

Entradas:  
Perfil de aceleraci√≥n a(t) en todo tiempo

```python
s0=0
v0=0
```

```python
delta_t = delta

all_v = np.zeros(n_steps+1)
all_s = np.zeros(n_steps+1)

v=v0
s=s0

for x in range(n_steps):
    
    s = s + v*delta_t
    v = v + a[x]*delta_t
    #
    
    all_v[x] = v
    all_s[x] = s
```

```python
plt.subplot(2,2,1)
plt.plot(t, a)
plt.title("Aceleracion")
plt.ylim(0,0.5)

plt.subplot(2,2,2)
plt.plot(t[:-1], all_v[:-1])
plt.title("Velocidad")

plt.subplot(2,2,3)
plt.plot(t[:-1], all_s[:-1])
plt.title("Posicion")
```
